```python
from darts_acquisition.s2.cdse_mosaic import CDSEMosaicStoreManager, _get_band_mapping, load_cdse_s2_mosaic
from darts_utils.paths import DefaultPaths
from darts_utils.paths import paths as darts_paths

defaults = DefaultPaths("...")
darts_paths.set_defaults(defaults)
cdse_mosaic_store = darts_paths.sentinel2_raw_data("cdse", "mosaics")
cdse_mosaic_store.mkdir(parents=True, exist_ok=True)
```

```python
s2id = "Sentinel-2_mosaic_2025_Q3_60WWS_0_0"
raw = CDSEMosaicStoreManager(
    store=cdse_mosaic_store, bands_mapping=_get_band_mapping("all"), aws_profile_name="default"
).open(s2id)
raw
```

```python
raw.B02[::10, ::10].plot(vmax=300)
```

```python
mosaic = load_cdse_s2_mosaic(s2id, store=cdse_mosaic_store, device="cpu")
mosaic
```

```python
mosaic.blue[::10, ::10].plot(vmax=0.1, vmin=0)
```

```python
mosaic.red[::10, ::10].plot(vmax=0.1, vmin=0)
```
